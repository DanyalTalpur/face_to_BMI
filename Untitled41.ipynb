```python
majority_class_folder = None
num_majority_class_images = -1
```

```python
for class_folder in class_folders:
    class_path = os.path.join(augmented_folder, class_folder)
    class_images = os.listdir(class_path)
    num_class_images = len(class_images)
    if num_class_images > num_majority_class_images:
        num_majority_class_images = num_class_images
        majority_class_folder = class_folder
```

```python
for minority_class_folder in class_folders:
    minority_class_path = os.path.join(augmented_folder, minority_class_folder)
    if not os.path.isdir(minority_class_path):
        continue 
    
    if minority_class_folder == majority_class_folder:
        continue
    
    minority_class_images = os.listdir(minority_class_path)
    
    oversampling_ratio = num_majority_class_images // len(minority_class_images)
    
    for i in range(oversampling_ratio):
        for minority_class_image in minority_class_images:
            output_subfolder_path = os.path.join(output_folder, minority_class_folder)
            
            if not os.path.exists(output_subfolder_path):
                os.makedirs(output_subfolder_path)
            
            minority_class_image_path = os.path.join(minority_class_path, minority_class_image)
            
            augmented_image_filename = os.path.splitext(minority_class_image)[0] + f'_augmented_{i+1}' + os.path.splitext(minority_class_image)[1]
            
            output_image_path = os.path.join(output_subfolder_path, augmented_image_filename)
            
            shutil.copyfile(minority_class_image_path, output_image_path)

    print(f'Oversampled images for {minority_class_folder}: {num_majority_class_images} saved in folder {minority_class_folder}')
    
print('Oversampling completed successfully.')
```

```python
data_dir = 'D:/BMIdataset/balancedData'
input_shape = (224, 224, 3)
num_classes = 1
batch_size = 32
epochs = 10
k = 5

images = []
labels = []
for folder in os.listdir(data_dir):
    folder_path = os.path.join(data_dir, folder)
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)
            if file.endswith('.jpg'):
                img = cv2.imread(file_path)
                img = cv2.resize(img, input_shape[:2])
                img = img.astype('float32') / 255.0
                images.append(img)
                label = None
                try:
                    label = float(folder)
                except ValueError:
                    print(f'Error extracting label from folder name: {folder}. Skipping...')
                    continue
                if label is not None:
                    labels.append(label)

images = np.array(images)
labels = np.array(labels)


base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)

model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(1, activation='linear'))

model.compile(optimizer=Adam(lr=0.001), loss='mean_squared_error', metrics=['mse'])

kf = KFold(n_splits=k, shuffle=True, random_state=42)
fold = 1
mse_scores = []
for train_index, test_index in kf.split(images):
    print(f'Fold: {fold}')
    X_train, X_test = images[train_index], images[test_index]
    y_train, y_test = labels[train_index], labels[test_index]

    datagen = ImageDataGenerator(rotation_range=20, width_shift_range=0.1,
                                 height_shift_range=0.1, horizontal_flip=True)
    datagen.fit(X_train)

    early_stop = EarlyStopping(monitor='val_loss', patience=3)
    history = model.fit_generator(datagen.flow(X_train, y_train, batch_size=batch_size),
                                  steps_per_epoch=len(X_train) // batch_size,
                                  epochs=epochs, validation_data=(X_test, y_test),
                                  callbacks=[early_stop])

    mse = model.evaluate(X_test, y_test, verbose=0)[1]
    mse_scores.append(mse)

    fold += 1

avg_mse = np.mean(mse_scores)
print(f'Average MSE: {avg_mse:.4f}')
```

```python
data_dir = 'D:/BMIdataset/balancedData'
input_shape = (224, 224, 3)
num_classes = 1
batch_size = 32
epochs = 10
k = 5

images = []
labels = []
for folder in os.listdir(data_dir):
    folder_path = os.path.join(data_dir, folder)
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)
            if file.endswith('.jpg'):
                img = cv2.imread(file_path)
                img = cv2.resize(img, input_shape[:2])
                img = img.astype('float32') / 255.0
                images.append(img)
                label = None
                try:
                    label = float(folder) 
                except ValueError:
                    print(f'Error extracting label from folder name: {folder}. Skipping...')
                    continue
                if label is not None:
                    labels.append(label)

images = np.array(images)
labels = np.array(labels)


model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape))
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(BatchNormalization())
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(BatchNormalization())
model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(BatchNormalization())
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(1, activation='linear'))

model.compile(optimizer=Adam(lr=0.001), loss='mean_squared_error', metrics=['mse'])

kf = KFold(n_splits=k, shuffle=True, random_state=42)
fold = 1
mse_scores = []
for train_index, test_index in kf.split(images):
    print(f'Fold: {fold}')
    X_train, X_test = images[train_index], images[test_index]
    y_train, y_test = labels[train_index], labels[test_index]

    datagen = ImageDataGenerator(rotation_range=20, width_shift_range=0.1,
                                 height_shift_range=0.1, horizontal_flip=True)
    datagen.fit(X_train)

    early_stop = EarlyStopping(monitor='val_loss', patience=3)
    history = model.fit_generator(datagen.flow(X_train, y_train, batch_size=batch_size),
                                  steps_per_epoch=len(X_train) // batch_size,
                                  epochs=epochs, validation_data=(X_test, y_test),
                                  callbacks=[early_stop])
    
    mse = model.evaluate(X_test, y_test, batch_size=batch_size)[1]
    mse_scores.append(mse)
    fold += 1

mse_avg = np.mean(mse_scores)
print('Average MSE:', mse_avg)
```
